````python
import numpy as np

def parse_input(input_string: str) -> list[np.array]:
    """Parse the input string into a list of numpy arrays"""
    key_locks = []
    for key_lock in input_string.strip().split('\n\n'):
        key_locks.append(np.array([list(row) for row in key_lock.split('\n')]))
    return key_locks

def update_array(key_lock: np.array) -> np.array:
    """Replace all . with 0 and # with 1"""
    return np.where(key_lock == '.', 0, 1)

def get_keys_locks(key_locks: list[np.array]) -> np.array:
    """Get all keys from the key_lock"""
    keys = []
    locks = []
    for key_lock in key_locks:
        if key_lock[0, 0] == 1:
            locks.append(key_lock)
        else:
            keys.append(key_lock)
    return keys, locks

def check_combinations(keys: list[np.array], locks: list[np.array]) -> int:
    """Check the number of compatible keys and locks"""
    compatible_count = 0
    for key in keys:
        for lock in locks:
            if np.all(key + lock <= 1):
                compatible_count += 1
    return compatible_count

with open('./data/Day 25/input.txt') as f:
    input_data = f.read()

key_locks = parse_input(input_data)
for i, key_lock in enumerate(key_locks):
    key_locks[i] = update_array(key_lock)

keys, locks = get_keys_locks(key_locks)

compatible_count = check_combinations(keys, locks)
print(compatible_count)
````
