```python
import networkx as nx

def build_graph(input_lines: list[str]) -> nx.DiGraph:
    G = nx.DiGraph()
    for line in input_lines:
        source, targets = line.split(":")
        source = source.strip()
        targets = [t.strip() for t in targets.split(" ") if t.strip()]
        for target in targets:
            G.add_edge(source, target)
    return G
```

```python
len(list(nx.all_simple_paths(build_graph(example), source="you", target="out")))
```

```python
len(list(nx.all_simple_paths(build_graph(data), source="you", target="out")))
```

```python
# Function to parse lines into a graph (adjacency list) and count all paths from source to target passing through required nodes
from functools import lru_cache

def parse_lines(input_lines: list[str]) -> dict:
    graph = {}
    for line in input_lines:
        source, targets = line.split(":")
        source = source.strip()
        targets = [t.strip() for t in targets.split(" ") if t.strip()]
        graph[source] = targets
    return graph

def count_paths_with_nodes(graph: dict, source: str, target: str, required_nodes: set) -> int:
    """
    Returns the count of all paths from source to target that pass through all required_nodes (at least once).
    """
    @lru_cache(maxsize=None)
    def count_paths(current, visited_fft=False, visited_dac=False):
        if current == "out":
            return visited_fft and visited_dac
        count = 0
        for neighbor in graph.get(current, []):
            count += count_paths(neighbor, visited_fft or neighbor == "fft", visited_dac or neighbor == "dac")
        return count
    return count_paths(source, source == "fft", source == "dac")
```

```python
count_paths_with_nodes(parse_lines(example2), "svr", "out", {"fft", "dac"})
```

```python
count_paths_with_nodes(parse_lines(data), "svr", "out", {"fft", "dac"})
```
