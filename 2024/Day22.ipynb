````python
def generate_secret_number(secret):
    secret = (secret * 64) ^ secret
    secret %= 16777216
    secret = (secret // 32) ^ secret
    secret %= 16777216
    secret = (secret * 2048) ^ secret
    secret %= 16777216
    return secret


def get_prices_and_changes(initial_secret):
    secret = initial_secret
    prices = []

    for _ in range(2001):  # 2000 changes need 2001 prices
        price = secret % 10
        prices.append(price)
        secret = generate_secret_number(secret)

    changes = [prices[i + 1] - prices[i] for i in range(len(prices) - 1)]

    return prices, changes


def build_change_dict(initial_secrets):
    """
    Build a dictionary that maps four consecutive changes to the price at which they occur.
    """
    change_dicts = []

    for secret in initial_secrets:
        prices, changes = get_prices_and_changes(secret)
        change_dict = {}

        for i in range(len(changes) - 3):
            change_seq = tuple(changes[i : i + 4])
            if change_seq not in change_dict:
                change_dict[change_seq] = prices[i + 4]

        change_dicts.append(change_dict)

    return change_dicts


def find_best_change_sequence(initial_secrets):
    possible_changes = range(-9, 10)

    change_dicts = build_change_dict(initial_secrets)

    max_bananas = 0
    best_sequence = None

    from itertools import product

    for sequence in product(possible_changes, repeat=4):
        total_bananas = 0
        for change_dict in change_dicts:
            if sequence in change_dict:
                total_bananas += change_dict[sequence]

        if total_bananas > max_bananas:
            max_bananas = total_bananas
            best_sequence = sequence

    return best_sequence, max_bananas


initial_secrets = [1, 2, 3, 2024]

best_sequence, max_bananas = find_best_change_sequence(initial_secrets)

print("Best Change Sequence:", best_sequence)
print("Maximum Bananas:", max_bananas)
````
