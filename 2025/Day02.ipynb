```python
example = """11-22,95-115,998-1012,1188511880-1188511890,222220-222224,
1698522-1698528,446443-446449,38593856-38593862,565653-565659,
824824821-824824827,2121212118-2121212124"""
```

```python
def split_input(text: str):
    text = text.replace("\n","")
    return text.split(",")

def get_range(row: str):
    items = row.split("-")
    return [int(i) for i in items]

def get_invalids(row: list[int]):
    invalids = 0
    for i in range(row[0], row[1]+1):
        textnr = str(i)
        if len(textnr) %2 != 0:
            continue
        half = len(textnr)//2
        if textnr[:half] == textnr[half:]:
            invalids += i
    return invalids
```

```python
invalid = 0
for i in split_input(example):
    row = get_range(i)
    invalid += get_invalids(row)
invalid
```

```python
invalid = 0
for i in split_input(data):
    row = get_range(i)
    invalid += get_invalids(row)
invalid
```

```python
from functools import cache


@cache
def find_factors(n):
    factors = []
    for i in range(1, abs(n) + 1):
        if n % i == 0:
            factors.append((i, n // i))
    return factors


def all_elements_same(lst):
    return all(x == lst[0] for x in lst) if lst else True


def get_invalids_2(row: list[int]):
    invalids = 0
    for i in range(row[0], row[1]+1):
        textnr = str(i)
        found = False
        for fact in find_factors(len(textnr)):
            if fact[0] == 1:
                continue
            temp = []
            for j in range(fact[0]):
                min1 = (j) * fact[1]
                max1 = (j+1) * fact[1]
                temp.append(textnr[min1:max1])
            if all_elements_same(temp) and not found:
                found = True
                invalids += i
                break
        found = False
    return invalids
```

```python
invalid = 0
for i in split_input(example):
    row = get_range(i)
    invalid += get_invalids_2(row)
invalid
```
