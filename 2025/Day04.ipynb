```python
example = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@."""
```

```python
import numpy as np
def parse_input(data: str) -> np.array:
    data = data.replace(".", "0")
    data = data.replace("@", "1")
    return np.array([[int(j) for j in i ] for i in data.splitlines()])

def count_rolls(paper_array: np.array) -> np.array:
    shape = np.shape(paper_array)
    output = np.zeros(shape)
    paper_array_padded = np.pad(paper_array, 1, "constant", constant_values=0)
    for j in range(1, shape[1]+1):
        for i in range(1, shape[0]+1):
            subsection = paper_array_padded[i - 1 : i + 2, j - 1 : j + 2]
            output[i-1][j-1] = np.sum(np.sum(subsection, dtype=np.int32), dtype=np.int32)
    return np.count_nonzero(np.where(output[np.where(paper_array == 1)]<=4))+1
```

```python
count_rolls(parse_input(example))
```

```python
count_rolls(parse_input(data))
```

```python
def removable_rolls(paper_array: np.array) -> np.array:
    shape = np.shape(paper_array)
    output = np.zeros(shape)
    paper_array_padded = np.pad(paper_array, 1, "constant", constant_values=0)
    for j in range(1, shape[1] + 1):
        for i in range(1, shape[0] + 1):
            subsection = paper_array_padded[i - 1 : i + 2, j - 1 : j + 2]
            output[i - 1][j - 1] = np.sum(
                np.sum(subsection, dtype=np.int32), dtype=np.int32
            )
    cond = output <= 4
    cond2  = paper_array == 1
    combined_cond = cond & cond2
    return np.argwhere(combined_cond)

def update_papers(arr, indices):
    for idx in indices:
        arr[tuple(idx)] = 0
    return arr

def take_rolls(paper_array: np.array, removable: int) -> tuple[np.array, int]:
    total_removed = 0
    while removable > 0:
        removelist = removable_rolls(paper_array)
        update_papers(paper_array, removelist)
        removable = len(removelist)
        total_removed += removable
    return total_removed
```

```python
take_rolls(parse_input(example),1)
```

```python
take_rolls(parse_input(data), 1)
```
