```python
import networkx as nx

def build_graph(input_lines: list[str]) -> nx.DiGraph:
    G = nx.DiGraph()
    for line in input_lines:
        source, targets = line.split(":")
        source = source.strip()
        targets = [t.strip() for t in targets.split(" ") if t.strip()]
        for target in targets:
            G.add_edge(source, target)
    return G
```

```python
len(list(nx.all_simple_paths(build_graph(example), source="you", target="out")))
```

```python
len(list(nx.all_simple_paths(build_graph(data), source="you", target="out")))
```

```python
from tqdm import tqdm

graph2 = build_graph(data)
passing_nodes = ["fft", "dac"]
source = "svr"
target = "out"
## Quick check order
# Check if paths exist between consecutive waypoints
for i in range(len(passing_nodes) + 1):
    if i == 0:
        src = source
        dst = passing_nodes[0]
    elif i == len(passing_nodes):
        src = passing_nodes[-1]
        dst = target
    else:
        src = passing_nodes[i-1]
        dst = passing_nodes[i]
    
    if nx.has_path(graph2, src, dst):
        shortest = nx.shortest_path(graph2, src, dst)
        print(f"{src} -> {dst}: ✓ (shortest: {' -> '.join(shortest)})")
    else:
        print(f"{src} -> {dst}: ✗ No path exists")

# Check both orderings
print("\nChecking order: svr -> fft -> dac -> out")
print(f"Has path: {nx.has_path(graph2, source, passing_nodes[0]) and nx.has_path(graph2, passing_nodes[0], passing_nodes[1]) and nx.has_path(graph2, passing_nodes[1], target)}")

print("\nChecking order: svr -> dac -> fft -> out")
print(f"Has path: {nx.has_path(graph2, source, passing_nodes[1]) and nx.has_path(graph2, passing_nodes[1], passing_nodes[0]) and nx.has_path(graph2, passing_nodes[0], target)}")

```
