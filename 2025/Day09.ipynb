```python
with open('./data/day9.txt', 'r') as f:
    data = f.read()
```

```python
example = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3"""
```

```python
def parse_input(input_string: str):
    return [[int(j) for j in i.split(",")] for i in input_string.splitlines()]

def largest_square(nodes):
    squares = []
    for i, node in enumerate(nodes):
        for j, node2 in enumerate(nodes):
            if i >= j:
                continue
            surface = (abs(node[0]-node2[0])+1) * (abs(node[1]-node2[1])+1)
            squares.append((i, j, surface))
    squares.sort(key=lambda x: x[2], reverse=True)
    return squares
```

```python
largest_square(parse_input(example))[0][2]
```

```python
largest_square(parse_input(data))[0][2]
```

```python
from shapely import Polygon

def largest_square_within_shape(nodes):
    squares = []
    outer_limit = Polygon(nodes)
    for i, node in enumerate(nodes):
        for j, node2 in enumerate(nodes):
            if i >= j:
                continue
            surface = (abs(node[0] - node2[0]) + 1) * (abs(node[1] - node2[1]) + 1)
            edges = [tuple(node), tuple([node2[0], node[1]]),tuple(node2), tuple([node[0], node2[1]])]
            if outer_limit.contains(Polygon(edges)):
                squares.append((i, j, surface))
    squares.sort(key=lambda x: x[2], reverse=True)
    return squares
```

```python
squares = largest_square_within_shape(parse_input(example))
squares[0][2]
```

```python
squares = largest_square_within_shape(parse_input(data))
squares[0][2]
```
