```python
with open('./data/day8.txt') as f:
    data = f.read()
```

```python
def parse_input(input_string):
    pointlist = []
    for line in input_string.strip().split("\n"):
        point = tuple(int(x) for x in line.split(","))
        pointlist.append(point)
    return pointlist


def calculate_square_distance(pointlist):
    output = []
    for i, point in enumerate(pointlist):
        for j, point2 in enumerate(pointlist):
            if i >= j:
                continue
            dist = sum((a - b) ** 2 for a, b in zip(point, point2))
            output.append((i, j, dist))
    return output

import networkx as nx

def create_graph(distances):
    G = nx.Graph()
    for i, j, dist in distances:
        G.add_edge(i, j, weight=dist)
    return G

def get_connected_components(G):
    components = list(nx.connected_components(G))
    components.sort(key=len, reverse=True)
    return components
```

```python
import math
distances = calculate_square_distance(parse_input(example))
shortest_distances = sorted(distances, key=lambda x: x[2])
G = create_graph(shortest_distances[:10])
circuits = get_connected_components(G)
math.prod(len(c) for c in circuits[:3])
```

```python
def create_graph(nodes):
    G = nx.Graph()
    G.add_nodes_from(nodes)
    return G

def update_graph(G, distance):
    i, j, dist = distance
    G.add_edge(i, j, weight=dist)
    return G
```

```python
import math
pointlist = parse_input(example)
distances = calculate_square_distance(pointlist)
shortest_distances = sorted(distances, key=lambda x: x[2])
G = create_graph(list(range(len(pointlist))))

while len(get_connected_components(G)) > 1:
    distance = shortest_distances.pop(0)
    G = update_graph(G, distance)

box_a = pointlist[distance[0]]
box_b = pointlist[distance[1]]

math.prod([box_a[0], box_b[0]])
```
