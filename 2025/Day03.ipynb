```python
example="""987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
with open("./data/day3.txt", "r") as f:
    data = f.read()
```

```python
def parse_input(data: str):
    return data.splitlines()

def find_max(batteries: str):
    batteries_list = [int(x) for x in batteries]
    max_range = batteries_list[:-1]
    i = batteries.find(str(max(max_range)))
    second_number = batteries_list[i+1:]
    j = batteries.find(str(max(second_number)))
    return int(batteries[i] + batteries[j])
```

```python
parsed_example = parse_input(example)
total = 0
for battery in parsed_example:
    total += find_max(battery)
print(total)
```

```python
parsed_data = parse_input(data)
total = 0
for battery in parsed_data:
    total += find_max(battery)
print(total)
```

```python
def find_max(batteries: str, nr_length: int, start: int):
    if nr_length == 0:
        return ""
    batteries_list = [int(x) for x in batteries]
    if -nr_length + 1 != 0:
        max_range = batteries_list[:-nr_length+1]
    else:
        max_range = batteries_list
    i = batteries.find(str(max(max_range)))
    return batteries[i] + find_max(batteries[i+1:], nr_length-1, i+1)
```
