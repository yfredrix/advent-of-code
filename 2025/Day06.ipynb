```python
example = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
```

```python
with open("./data/day6.txt") as f:
    data = f.read()
```

```python
import math


def parse_input(data):
    lines = data.splitlines()
    math_problem = {}
    operator = []
    for line in lines:
        problems = line.split()
        for i, problem in enumerate(problems):
            if i not in math_problem:
                math_problem[i] = []
            if problem == '*':
                operator.append('*')
            elif problem == '+':
                operator.append('+')
            else:
                math_problem[i].append(int(problem))
    return math_problem, operator

def solve_math(math_problem, operator):
    results = []
    for i in range(len(operator)):
        if operator[i] == '*':
            results.append(math.prod(math_problem[i]))
        elif operator[i] == '+':
            results.append(sum(math_problem[i]))
    return results
```

```python
sum(solve_math(*parse_input(example)))
```

```python
sum(solve_math(*parse_input(data)))
```

```python
def check_spaces(i ,lines):
    for line in lines:
        if line[i] != ' ':
            return False
    return True

def parse_input_part2(data):
    lines = data.splitlines()
    math_problem = {}
    operator = []
    problem_index = 0
    for i, _ in enumerate(lines[0]):
        if problem_index not in math_problem:
            math_problem[problem_index] = []
        num_str = ""
        next_line = False
        for line in lines:
            if next_line:
                break
            index = len(line)-i-1
            if index >= len(line):
                continue
            char = line[index]
            if char == " ":
                if check_spaces(index, lines):
                    problem_index += 1
                    next_line = True
                    continue
                else:
                    continue
            if char == '*':
                operator.append('*')
            elif char == '+':
                operator.append('+')
            else:
                num_str += char
        if num_str != "":
            math_problem[problem_index].append(int(num_str))
    return math_problem, operator
```

```python
sum(solve_math(*parse_input_part2(example)))
```

```python
sum(solve_math(*parse_input_part2(data)))
```
