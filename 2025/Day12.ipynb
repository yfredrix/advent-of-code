```python
with open("./data/example12.txt", "r") as f:
    example = f.read()

with open("./data/example_presents_12.txt", "r") as f:
    example_presents_raw = f.read()
```

```python
with open("./data/day12.txt", "r") as f:
    data = f.read()

with open("./data/presents12.txt", "r") as f:
    presents_raw = f.read()
```

```python
import math
from functools import cache

def parse_presents(raw):
    presents = {}
    for line in raw.splitlines():
        if ":" in line:
            key = int(line[0])
            if key not in presents:
                presents[key] = []
        else:
            if line != "":
                presents[key].append(list(line))
    return presents

def parse_challenge(raw):
    challenges = []
    for line in raw.splitlines():
        grid, instructions = line.split(":")
        area = math.prod(int(x) for x in grid.split("x"))
        instructions = [int(x) for x in instructions.strip().split(" ")]
        challenges.append((area, instructions))
    return challenges

def present_size(present):
    size = 0
    for row in present:
        for cell in row:
            if cell == "#":
                size += 1
    return size
```

```python
example_presents = parse_presents(example_presents_raw)
example_challenges = parse_challenge(example)
example_present_sizes = {}
for key in example_presents.keys():
    example_present_sizes[key] = present_size(example_presents[key])

counter = 0
for area, instructions in example_challenges:
    total_present_size = 0
    for key, nr in enumerate(instructions):
        total_present_size += nr * example_present_sizes[key]
    if total_present_size < area:
        counter += 1
print(counter)
```

```python
presents = parse_presents(presents_raw)
challenges = parse_challenge(data)
present_sizes = {}
for key in presents.keys():
    present_sizes[key] = present_size(presents[key])
counter = 0
for area, instructions in challenges:
    total_present_size = 0
    for key, nr in enumerate(instructions):
        total_present_size += nr * present_sizes[key]
    if total_present_size < area:
        counter += 1
print(counter)
```
