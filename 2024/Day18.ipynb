````python
from collections import deque


def shortest_path_in_memory_space(grid_size, incoming_positions, num_bytes):
    # Initialize the memory grid
    grid = [["." for _ in range(grid_size)] for _ in range(grid_size)]

    # Simulate the first `num_bytes` falling onto the grid
    for y, x in incoming_positions[:num_bytes]:
        # Ensure the coordinates are swapped for grid indexing
        grid[x][y] = "#"

    # Directions for up, down, left, right
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    def bfs_shortest_path(start, goal):
        queue = deque([(start, 0)])  # (position, step_count)
        visited = set([start])

        while queue:
            (current_x, current_y), steps = queue.popleft()

            if (current_x, current_y) == goal:
                return steps

            for dx, dy in directions:
                new_x, new_y = current_x + dx, current_y + dy

                if 0 <= new_x < grid_size and 0 <= new_y < grid_size:
                    if grid[new_y][new_x] == "." and (new_x, new_y) not in visited:
                        visited.add((new_x, new_y))
                        queue.append(((new_x, new_y), steps + 1))

        return float("inf")  # No path found

    # Start and goal positions
    start = (0, 0)
    goal = (grid_size - 1, grid_size - 1)

    # Calculate the shortest path
    min_steps = bfs_shortest_path(start, goal)
    return min_steps


# Example scenario: 7x7 grid, first 12 positions
example_incoming_positions = [
    (5, 4),
    (4, 2),
    (4, 5),
    (3, 0),
    (2, 1),
    (6, 3),
    (2, 4),
    (1, 5),
    (0, 6),
    (3, 3),
    (2, 6),
    (5, 1),
    (1, 2),
    (5, 5),
    (2, 5),
    (6, 5),
    (1, 4),
    (0, 4),
    (6, 4),
    (1, 1),
    (6, 1),
    (1, 0),
    (0, 5),
    (1, 6),
    (2, 0),
]

# Grid size of 7x7, considering the first 12 bytes
result = shortest_path_in_memory_space(7, example_incoming_positions, 12)
print(f"The minimum number of steps needed to reach the exit is: {result}")
````
