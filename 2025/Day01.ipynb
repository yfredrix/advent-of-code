```python
with open("./data/day1.txt", "r") as f:
    data = f.read().splitlines()
```

```python
def turn_dail(cur_pos: int, direction: str, nr_places: int):
    if direction == "L":
        nr_places *= -1
    if cur_pos + nr_places < 0:
        temp = cur_pos + nr_places
        return (100 + temp) % 100
    return (cur_pos + nr_places) % 100
```

```python
def split_input(inputdata: list[str]):
    cleared_output = []
    for line in inputdata:
        cleared_output.append((line[0], int(line[1:])))
    return cleared_output
```

```python
example = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82""".splitlines()
```

```python
example_in = split_input(example)
position = 50
nr = 0
for i in example_in:
    position = turn_dail(position, i[0], i[1])
    if position == 0 or position == 100:
        nr += 1
print(nr)
```

```python
data_in = split_input(data)
position = 50
nr = 0
for i in data_in:
    position = turn_dail(position, i[0], i[1])
    if position == 0:
        nr += 1
print(nr)
```

```python
def turn_dial2(cur_pos: int, direction: str, nr_places: int):
    zero_count = 0
    for _ in range(nr_places):
        if direction == "L":
            cur_pos = (cur_pos - 1) %100
        else:
            cur_pos = (cur_pos + 1) %100
        if cur_pos == 0:
            zero_count += 1
    return cur_pos, zero_count
```

```python
example_in = split_input(example)
position = 50
nr = 0
for i in example_in:
    position, tempnr = turn_dial2(position, i[0], i[1])
    nr += tempnr
print(nr)
```

```python
position = 50
nr = 0
for i in data_in:
    position, tempnr = turn_dial2(position, i[0], i[1])
    nr += tempnr
print(nr)
```
