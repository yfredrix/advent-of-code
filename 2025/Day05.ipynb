```python
example = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""
```

```python
with open("./data/day5.txt") as f:
    puzzle_input = f.read()
```

```python
def create_input(input_string):
    sections = input_string.strip().split("\n\n")
    fresh_ingredients = []
    for line in sections[0].splitlines():
        start, end = map(int, line.split("-"))
        fresh_ingredients.append((start, end))
    ingredients = [int(line) for line in sections[1].splitlines()]
    return fresh_ingredients, ingredients
```

```python
create_input(example)
```

```python
def check_valid_ingredient(fresh_ingredients, ingredient):
    for start, end in fresh_ingredients:
        if start <= ingredient <= end:
            return True
    return False
```

```python
example_fresh, example_ingredient = create_input(example)
available = 0
for ingredient in example_ingredient:
    available += check_valid_ingredient(example_fresh, ingredient)
available
```

```python
fresh, ingredient = create_input(puzzle_input)
available = 0
for item in ingredient:
    available += check_valid_ingredient(fresh, item)
available
```

```python
def total_fresh_unique_ingredients(fresh_ingredients: list[tuple[int, int]]):
    fresh_ingredients.sort(key=lambda x: x[0])
    unique_ingredients = list()
    for start, end in fresh_ingredients:
        if not any(ingredient["min"] <= start <= ingredient["max"] or ingredient["min"] <= end <= ingredient["max"] for ingredient in unique_ingredients):
            unique_ingredients.append({"min": start, "max": end})
        if any(ingredient["min"] <= start <= ingredient["max"] for ingredient in unique_ingredients):
            for ingredient in unique_ingredients:
                if (
                    end > ingredient["max"]
                    and ingredient["min"] <= start <= ingredient["max"]
                ):
                    ingredient["max"] = end
    total_fresh = 0
    for ingredient in unique_ingredients:
        ingredient["total"] = ingredient["max"] - ingredient["min"] + 1
        total_fresh += ingredient["total"]
    return unique_ingredients, total_fresh
```
